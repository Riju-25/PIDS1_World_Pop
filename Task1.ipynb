Importing Libraries and Reading Dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
data = pd.read_csv('pop_data.csv')
data.head()
```

```python
data.isna().sum()
```

Removing unnecessary Columns

```python
data.drop(['Series Name', 'Country Name'], axis=1, inplace=True)
```

Extracting Top 10 countries in year 2022

```python
ttl_pop=data[data['Series Code']=='SP.POP.TOTL']
sorted1=ttl_pop.sort_values(by='2022', ascending=False)
top_10=sorted1.head(10)
print("Top 10 Countries with Highest Population in 2022:")
print(top_10['Country Code'])
```

Bar Plot for highest populated countries in year 2022

```python
import seaborn as sns
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='Country Code', y='2022', data=top_10, hue='Country Code', palette = "seismic")
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.title("Top 10 populated countries in year 2022", fontsize=15)
plt.show()
```

Extracting least 10 populated countries in year 2022

```python
sorted2=ttl_pop.sort_values(by='2022', ascending=True)
bottom_10=sorted2.head(10)
print("Bottom 10 Countries with Highest Population in 2022:")
print(bottom_10['Country Code'])
```

Bar Plot for least populated countries in year 2022

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='Country Code', y='2022', data=bottom_10, hue='Country Code', palette = "seismic_r")
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.title("Least 10 populated countries in year 2022", fontsize=15)
plt.show()
```

Extracting Top 10 Male populated countries in year 2020

```python
ttl_pop_male=data[data['Series Code']=='SP.POP.TOTL.MA.IN']
sorted3=ttl_pop_male.sort_values(by='2020', ascending=False)
top_male_10=sorted3.head(10)
print("Top 10 Countries with Highest Male Population in 2020:")
print(top_male_10['Country Code'])
```

Bar Plot for top male populated Countries in year 2020

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='Country Code', y='2020', data=top_male_10, hue='Country Code', palette = "seismic")
plt.xlabel("Male Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.title("Top 10 Male populated countries in year 2020", fontsize=15)
plt.show()
```

Extracting Least 10 Female populated Countries in year 2018

```python
ttl_pop_female=data[data['Series Code']=='SP.POP.TOTL.FE.IN']
sorted4=ttl_pop_male.sort_values(by='2018', ascending=True)
bottom_female_10=sorted4.head(10)
print("Top 10 Countries with least Female Population in 2018:")
print(bottom_female_10['Country Code'])
```

Bar Plot for least 10 Female populated Countries in year 2018

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='Country Code', y='2018', data=bottom_female_10, hue='Country Code', palette = "seismic_r")
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.title("Least 10 Female populated countries in year 2018", fontsize=15)
plt.show()
```

Merging Male and Female population data

```python
merge_data=pd.merge(ttl_pop_male,ttl_pop_female ,on="Country Code",suffixes=("_male","_female"))
merge_data.head()
```

Forming total population column for year 2016

```python
merge_data["Total population 2016"] = merge_data["2016_male"] + merge_data ["2016_female"]
```

Extracting top 10 populated countries in year 2016

```python
sorted_data = merge_data.sort_values(by="Total population 2016", ascending=False)
top_10_countries = sorted_data.head(10)
```

Stacked Bar plot for top 10 highest populated Countries in year 2016

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="Country Code",y="2016_female",data=top_10_countries,color="maroon",label="Female Population")
sns.barplot(x="Country Code",y="2016_male",data=top_10_countries,bottom=top_10_countries["2016_female"],color="blue",label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries with Highest Population in 2016 with Gender Distribution", fontsize=15)
plt.xticks(rotation=45,ha="right")
plt.show()
```
